```python
## groupby()
```

```python
np.random.seed(0)
df5 = pd.DataFrame({"key1": ["A", "A", "B", "B", "A"],
                    "key2": ["one", "two", "one", "two", "one"],
                    "data1": [1, 2, 3, 4, 5],
                    "data2": [10, 20, 30, 40, 50]})
df5
```

```python
groups = df5.groupby(df5.key1)
groups
```

```python
groups.groups  # key1 기준으로 같은 것끼리 묶음. {"Key": [index]}
```

```python
groups.sum()
```

```python
df5.data1.groupby(df5.key1).sum()
```

```python
df5.groupby(df5.key1)["data1"].sum()  # 같은 결과
```

```python
df5.groupby(df5.key1).sum()["data1"]  # 같은 결과
```

```python
## 복합 키(key1, key2) 값에 따른 data1의 합계
```

```python
df5.data1.groupby([df5.key1, df5.key2]).sum()  # key가 되는 column 2개 지정
```

```python
def change_boolean(value):
    if value == "yes":  # if절에 조건이 아니라 value를 넣는다
        return True
    else:
        return False

titanic_Q1 = titanic.alive.apply(change_boolean)
titanic_Q1.unique()
```
