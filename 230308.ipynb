```python
class NBATeam:
    def __init__(self, name, team):
        self.name = name
        self.team = team
        self.position = ""
        self.birthday = ""
        self.salary = 0
        
    def information(self):
        print(f"{self.name}의 팀은 {self.team} 입니다.")
```

```python
##### Data frame 고급 인덱싱 - loc 인덱서
```

```python
## df.loc[row 인덱스] 또는 df.loc[row 인덱스, column 인덱스]
```

```python
df = pd.DataFrame(np.arange(10, 22).reshape(3, 4),
                   index = ["a", "b", "c"],
                   columns = ["A", "B", "C", "D"])
df
```

```python
df.loc["a"]  # 인덱싱
```

```python
df.loc["b":"c"]  # 슬라이싱
```

```python
df["b":"c"]
```

```python
df.loc[["b", "c"]]
```

```python
df.A > 15
# a는 조건 불일치 -> False
# b는 조건 불일치 -> False
# c는 조건 일치 -> True
```

```python
df.loc[df.A > 15]  # True인 결과 출력!
```

```python
def select_rows(df, num):
    return df.A > num
```

```python
select_rows(df, 10)
```

```python
df.loc[select_rows(df, 10)]
```

```python
df.loc["A"]  # KeyError 발생. Column label 인덱싱은 불가능
```

```python
df.loc[["A", "B"]]  # KeyError 발생. 같은 이유
```

```python
## 참고 ##
# column label 인덱싱, column label 리스트 인덱싱은 불가능
# [0] 같은 index -> iloc[]
# "a" 같은 label -> loc[]
```

```python
df2 = pd.DataFrame(np.arange(10, 26).reshape(4, 4), columns = ["A", "B", "C", "D"])
df2
```

```python
df2.loc[1:2]  # 행 index값이 정수이면 마지막 값이 포함됨
```

```python
df.loc["a", "A"]
```

```python
df.at["a", "A"]

# df.loc[ ]와 가져오는 값은 같음
# 하나의 값에만 접근하므로 loc()보다 빠름
# 단, 슬라이싱은 불가능
```

```python
df.loc["b":, "A"]
```

```python
df.loc["a", :]
```

```python
df.loc[["a", "b"], ["B", "D"]]
```

```python
df.loc[df.A > 10, ["C", "D"]]
```
